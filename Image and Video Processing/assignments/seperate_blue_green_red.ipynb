```python
# Question 1
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
image = cv2.imread("image.jpg")
blue_channel, green_channel, red_channel = cv2.split(image)
empty_channel = np.zeros_like(blue_channel)

blue_grayscale = cv2.merge((blue_channel, empty_channel, empty_channel))
green_grayscale = cv2.merge((empty_channel, green_channel, empty_channel))
red_grayscale = cv2.merge((empty_channel, empty_channel, red_channel))

cv2_imshow( blue_grayscale)
cv2_imshow(green_grayscale)
cv2_imshow( red_grayscale)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Question 2
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread("image.jpg")

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2_imshow(gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
