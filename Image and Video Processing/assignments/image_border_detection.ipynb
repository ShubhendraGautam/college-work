```python
from PIL import Image, ImageEnhance
import cv2
from google.colab.patches import cv2_imshow

def enhance_image(image_path, output_path):
    # Open the image using PIL
    original_image = Image.open(image_path)

    # Enhance brightness
    brightness_enhancer = ImageEnhance.Brightness(original_image)
    enhanced_brightness = brightness_enhancer.enhance(1.5)  # Increase brightness by 50%

    # Enhance contrast
    contrast_enhancer = ImageEnhance.Contrast(enhanced_brightness)
    enhanced_contrast = contrast_enhancer.enhance(1.3)  # Increase contrast by 30%

    # Enhance sharpness
    sharpness_enhancer = ImageEnhance.Sharpness(enhanced_contrast)
    enhanced_sharpness = sharpness_enhancer.enhance(2.0)  # Increase sharpness

    # Enhance color
    color_enhancer = ImageEnhance.Color(enhanced_sharpness)
    enhanced_color = color_enhancer.enhance(1.2)  # Increase color saturation by 20%

    # Save the enhanced image
    enhanced_color.save(output_path)

if __name__ == "__main__":
    input_image_path = '/content/image.png'  # Replace with your image path
    output_image_path = 'enhanced_image.png'

    enhance_image(input_image_path, output_image_path)
    to_print=cv2.imread(output_image_path)
    cv2_imshow(to_print)
    print("Image enhancement complete. Enhanced image saved as 'enhanced_image.jpg'")
```

```python
from PIL import Image, ImageFilter

def perform_edge_detection(input_path, output_path):
    # Open the image using PIL
    original_image = Image.open(input_path)

    # Perform edge detection using the FIND_EDGES filter
    edge_detected_image = original_image.filter(ImageFilter.FIND_EDGES)

    # Save the edge-detected image
    edge_detected_image.save(output_path)

if __name__ == "__main__":
    input_image_path = 'image.png'  # Replace with your image path
    output_image_path = 'edge_detected_image.png'

    perform_edge_detection(input_image_path, output_image_path)

    to_print=cv2.imread(output_image_path)
    cv2_imshow(to_print)
    print("Edge detection complete. Edge-detected image saved as 'edge_detected_image.jpg'")
```
