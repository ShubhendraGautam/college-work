```python
import cv2
from google.colab.patches import cv2_imshow

# Load the image
image_path = "image.png"
image = cv2.imread(image_path)

# Get the original dimensions
original_height, original_width = image.shape[:2]

# Calculate new dimensions
new_width = int(original_width * 2)
new_height = int(original_height * 2)

# Resize the image
upscaled_image = cv2.resize(image, (new_width, new_height), interpolation=cv2.INTER_LINEAR)

# Save the upscaled image (optional)
cv2.imwrite("upscaled_image.png", upscaled_image)

# Display the original and upscaled images (optional)
print("Original Image")
cv2_imshow( image)
print("Upsacled Image")
cv2_imshow( upscaled_image)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load the image in grayscale (optional, depending on your image)
image_path = "image.png"
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Convert image to float32 and scale to the range [0, 1]
image_float = image.astype(np.float32) / 255.0

# Apply the log transformation
c = 1 / np.log(1 + np.max(image_float))  # Scaling constant
log_image = c * np.log(1 + image_float)

# Scale the log transformed image back to range [0, 255]
log_image_scaled = ((log_image - np.min(log_image)) / (np.max(log_image) - np.min(log_image))) * 255
log_image_8bit = np.uint8(log_image_scaled)

# Save the transformed image (optional)
cv2.imwrite("log_transformed_image.png", log_image_8bit)

# Display the original and transformed images (optional)
cv2_imshow( image)
cv2_imshow( log_image_8bit)

```
