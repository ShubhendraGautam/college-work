```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image_path = '/content/sddefault.jpg'
original_image = cv2.imread(image_path)

# Define gamma values
gamma_values = [0.25, 1, 2.5]

# Apply gamma transform and display images
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
plt.imshow(original_image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

for idx, gamma in enumerate(gamma_values, start=1):
    gamma_transformed = np.power(original_image/255.0, gamma)
    gamma_transformed = np.uint8(gamma_transformed * 255)

    plt.subplot(2, 2, idx+1)
    plt.imshow(gamma_transformed, cmap='gray')
    plt.title(f'Gamma = {gamma}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image_path = '/content/sddefault.jpg'
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Extract bit planes and display
plt.figure(figsize=(10, 8))

for i in range(8):
    bit_plane = (original_image >> i) & 1  # Extract ith bit
    bit_plane *= 255  # Scale to 0-255 range
    bit_plane = np.uint8(bit_plane)  # Convert to uint8 for display

    plt.subplot(3, 3, i + 1)
    plt.imshow(bit_plane, cmap='gray')
    plt.title(f'Bit Plane {i}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image_path = '/content/sddefault.jpg'
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Select the number of bit planes for reconstruction (e.g., 4)
n_bit_planes = 8

# Extract selected bit planes and combine them
reconstructed_image = np.zeros_like(original_image, dtype=np.uint8)

for i in range(n_bit_planes):
    bit_plane = (original_image >> i) & 1
    bit_plane *= (2 ** i)
    reconstructed_image += bit_plane

# Display the reconstructed image
plt.figure(figsize=(6, 6))
plt.imshow(reconstructed_image, cmap='gray')
plt.title(f'Reconstructed Image using {n_bit_planes} Bit Planes')
plt.axis('off')
plt.show()
```
