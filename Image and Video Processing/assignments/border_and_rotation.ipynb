```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow


image_path = '/content/image.png'
image = cv2.imread(image_path)

border_color = (0, 165, 255)
border_thickness = 10


new_height = image.shape[0] + 2 * border_thickness
new_width = image.shape[1] + 2 * border_thickness
bordered_image = np.full((new_height, new_width, 3), border_color, dtype=np.uint8)


bordered_image[border_thickness:border_thickness+image.shape[0], border_thickness:border_thickness+image.shape[1]] = image

cv2_imshow( bordered_image)




```

```python



image_path = '/content/image.png'
image = cv2.imread(image_path)

# Rotate clockwise by 90 degrees
rotated_clockwise = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)

# Rotate counterclockwise by 90 degrees
rotated_counterclockwise = cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE)

# Display the original and rotated images
cv2_imshow( image)
cv2_imshow( rotated_clockwise)
cv2_imshow( rotated_counterclockwise)
```
