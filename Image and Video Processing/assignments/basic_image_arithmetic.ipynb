```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

def load_image(image_path):
    # Load the image using OpenCV
    image = cv2.imread(image_path)
    return image

def display_image(image, window_name="Image"):
    # Display the image using OpenCV
    cv2_imshow(image)


def resize_image(image, width, height):
    # Resize the image to the given width and height
    return cv2.resize(image, (width, height))

def add_images(image1, image2):
    # Ensure the images have the same dimensions
    if image1.shape != image2.shape:
        raise ValueError("Images must have the same dimensions.")

    # Perform addition of the two images (pixel-wise)
    result = np.clip(image1 + image2, 0, 255).astype(np.uint8)
    return result

def subtract_images(image1, image2):
    # Ensure the images have the same dimensions
    if image1.shape != image2.shape:
        raise ValueError("Images must have the same dimensions.")

    # Perform subtraction of the two images (pixel-wise)
    result = np.clip(image1 - image2, 0, 255).astype(np.uint8)
    return result

def multiply_images(image1, image2):
    # Ensure the images have the same dimensions
    if image1.shape != image2.shape:
        raise ValueError("Images must have the same dimensions.")

    # Perform multiplication of the two images (pixel-wise)
    result = np.clip(image1 * image2, 0, 255).astype(np.uint8)
    return result

def divide_images(image1, image2):
    # Ensure the images have the same dimensions
    if image1.shape != image2.shape:
        raise ValueError("Images must have the same dimensions.")

    # Perform division of the two images (pixel-wise)
    # Avoid division by zero by adding a small epsilon value
    epsilon = 1e-5
    result = np.clip(image1 / (image2 + epsilon), 0, 255).astype(np.uint8)
    return result

# Load two images
image1 = load_image("img1.jpeg")
image2 = load_image("img2.jpeg")

# Resize images to the same dimensions (you can choose any desired dimensions)
width, height = 400, 300
image1 = resize_image(image1, width, height)
image2 = resize_image(image2, width, height)

# Perform addition
add_result = add_images(image1, image2)
display_image(add_result, "Addition")

# Perform subtraction
subtract_result = subtract_images(image1, image2)
display_image(subtract_result, "Subtraction")

# Perform multiplication
multiply_result = multiply_images(image1, image2)
display_image(multiply_result, "Multiplication")

# Perform division
divide_result = divide_images(image1, image2)
display_image(divide_result, "Division")
```
