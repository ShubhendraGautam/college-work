```python
# Question 1
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

# Load the image
i = cv2.imread('image.jpg')

# Convert the BGR image to RGB (OpenCV reads images in BGR format)
i = cv2.cvtColor(i, cv2.COLOR_BGR2RGB)

# Create a 2x2 subplot grid
plt.subplot(3, 2, 1)
plt.imshow(i)
plt.title('Original Image')

# Extract Red, Green, and Blue components
r, g, b = cv2.split(i)

# Red Component
plt.subplot(3, 2, 2)
plt.imshow(r, cmap='gray')
plt.title('Red Component')

# Green Component
plt.subplot(3, 2, 3)
plt.imshow(g, cmap='gray')
plt.title('Green Component')

# Blue Component
plt.subplot(3, 2, 4)
plt.imshow(b, cmap='gray')
plt.title('Blue Component')

# Convert to grayscale using OpenCV's cvtColor function
gray = cv2.cvtColor(i, cv2.COLOR_RGB2GRAY)

# Grayscale Image
plt.subplot(3, 2, 5)
plt.imshow(gray, cmap='gray')
plt.title('Gray Image')

# Adjust layout to avoid overlapping titles and labels
plt.tight_layout()

# Show the plot
plt.show()

```

```python
# Question 2
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the color image
I = cv2.imread('image.jpg')

# Convert BGR image to RGB
I_rgb = cv2.cvtColor(I, cv2.COLOR_BGR2RGB)

# Create a 2x2 subplot grid for the color image and its complement
plt.subplot(2, 2, 1)
plt.imshow(I_rgb)
plt.title('Color Image')

# Find the complement of the color image
c = cv2.bitwise_not(I)

# Convert BGR image to RGB
c_rgb = cv2.cvtColor(c, cv2.COLOR_BGR2RGB)

plt.subplot(2, 2, 2)
plt.imshow(c_rgb)
plt.title('Complement of Color Image')

# Convert the color image to grayscale
r = cv2.cvtColor(I_rgb, cv2.COLOR_RGB2GRAY)

plt.subplot(2, 2, 3)
plt.imshow(r, cmap='gray')
plt.title('Grayscale of Color Image')

# Find the complement of the grayscale image
b = cv2.bitwise_not(r)

plt.subplot(2, 2, 4)
plt.imshow(b, cmap='gray')
plt.title('Complement of Gray Image')

# Simulation of an Image (Arithmetic & Logic Operation)
a = np.ones((40, 40))
b = np.zeros((40, 40))
c = np.hstack((a, b))
d = np.hstack((b, a))
image_A = 10 * (c + d)
image_M = c * d
image_S = c - d
image_D = c / 4

# Create a 3x2 subplot grid for the simulation images
plt.figure()
plt.subplot(3, 2, 1)
plt.imshow(c, cmap='gray')

plt.subplot(3, 2, 2)
plt.imshow(d, cmap='gray')

plt.subplot(3, 2, 3)
plt.imshow(image_A, cmap='gray')

plt.subplot(3, 2, 4)
plt.imshow(image_M, cmap='gray')

plt.subplot(3, 2, 5)
plt.imshow(image_S, cmap='gray')

plt.subplot(3, 2, 6)
plt.imshow(image_D, cmap='gray')

plt.show()
```
